Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
query = "SELECT * FROM master"
df = pd.read_sql(query,conn)
```

```python
conn.close()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df["ConvertedComp"],log=True,bins=20)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import seaborn as sns
```

```python
import numpy as np
```

```python
sns.boxplot(data=df,x="Age")
plt.xticks(np.linspace(0,100,11))
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
sns.scatterplot(data=df,x="Age",y="WorkWeekHrs")
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
sns.scatterplot(data=df,x="WorkWeekHrs",y="CodeRevHrs",size="Age",sizes=(50, 300),alpha=0.6)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

```python
# your code goes here

query = "SELECT * FROM DatabaseDesireNextYear"
database_df = pd.read_sql(query,conn)
```

```python
pie_series = database_df.DatabaseDesireNextYear.value_counts().head()
plt.pie(pie_series.values,labels=pie_series.index,autopct='%1.1f%%')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
bar_df
```

```python
# your code goes here
bar_df = df[(df.Age >= 30) & (df.Age <= 35)].groupby("Age")[["WorkWeekHrs","CodeRevHrs"]].median()
bar_df.plot(kind="bar",stacked=True)
plt.title("Median Weekly Work Hours and Code Review Hours for the Age Group 30 to 35")
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
line_df = df.groupby("Age")["ConvertedComp"].median()
line_df.loc[45:60].plot(kind="line")
plt.title("Median Converted Compensations for All Ages From 45 to 60")
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
sns.countplot(data=df,x="MainBranch")
plt.xticks(rotation=25)
plt.title("Count of")
plt.show()
```
